```python
from google.colab import files

uploaded = files.upload()
```

```python
import hashlib
import shutil
import os

# Get uploaded file name
original_file = list(uploaded.keys())[0]

# Name of forensic copy
copy_file = "forensic_copy_" + original_file

# Create bit-stream (binary) copy
shutil.copyfile(original_file, copy_file)

# Function to calculate SHA-256 hash
def calculate_hash(filename):
    sha256 = hashlib.sha256()

    with open(filename, "rb") as file:
        while True:
            data = file.read(4096)
            if not data:
                break
            sha256.update(data)

    return sha256.hexdigest()

# Calculate hashes
original_hash = calculate_hash(original_file)
copy_hash = calculate_hash(copy_file)

print("=" * 50)
print("FORENSIC BIT-STREAM COPY REPORT")
print("=" * 50)

print("Original File :", original_file)
print("Copied File   :", copy_file)

print("\nOriginal SHA-256 Hash:")
print(original_hash)

print("\nCopied SHA-256 Hash:")
print(copy_hash)

print("\nIntegrity Verification:")

if original_hash == copy_hash:
    print("SUCCESS: Hashes Match")
    print("Integrity Verified")
else:
    print("FAILED: Hashes Do Not Match")
```

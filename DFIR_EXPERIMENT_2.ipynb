```python
# Create a sample evidence file

filename = "evidence.txt"

with open(filename, "w") as file:
    file.write("Digital Forensics Evidence")

print("Evidence file created successfully.")
```

```python
import hashlib

def calculate_hash(filename):
    sha256 = hashlib.sha256()

    with open(filename, "rb") as file:
        while True:
            data = file.read(4096)
            if not data:
                break
            sha256.update(data)

    return sha256.hexdigest()

original_hash = calculate_hash(filename)

print("Original SHA-256 Hash:")
print(original_hash)
```

```python
current_hash = calculate_hash(filename)

print("Current SHA-256 Hash:")
print(current_hash)

if original_hash == current_hash:
    print("\nEvidence Verified")
    print("Chain of Custody Maintained")
else:
    print("\nEvidence Modified")
    print("Chain of Custody Broken")
```

```python
with open(filename, "a") as file:
    file.write("\nEvidence was modified.")

print("Evidence file modified.")
```

```python
new_hash = calculate_hash(filename)

print("New SHA-256 Hash:")
print(new_hash)

if original_hash == new_hash:
    print("\nEvidence Verified")
    print("Chain of Custody Maintained")
else:
    print("\nEvidence Modified")
    print("Chain of Custody Broken")
```

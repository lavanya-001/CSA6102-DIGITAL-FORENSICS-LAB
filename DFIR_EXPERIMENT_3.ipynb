```python
# Function to detect phishing emails

def detect_phishing(sender, subject, body):

    suspicious_words = [
        "urgent",
        "verify",
        "password",
        "bank",
        "click",
        "login",
        "winner",
        "free",
        "account",
        "limited",
        "update"
    ]

    score = 0

    # Check sender
    if "@gmail.com" not in sender and "@outlook.com" not in sender:
        score += 1

    # Check subject
    for word in suspicious_words:
        if word.lower() in subject.lower():
            score += 1

    # Check body
    for word in suspicious_words:
        if word.lower() in body.lower():
            score += 1

    print("Sender :", sender)
    print("Subject:", subject)

    if score >= 3:
        print("\nResult: Phishing Email Detected")
    else:
        print("\nResult: Safe Email")
```

```python
sender = "support@fakebank.xyz"

subject = "Urgent! Verify Your Bank Account"

body = """
Dear Customer,

Your account has been suspended.

Click here immediately to verify your password.

Limited time.

"""

detect_phishing(sender, subject, body)
```

```python
sender = "teacher@gmail.com"

subject = "Assignment Submission"

body = """
Hello Student,

Please submit your assignment before Friday.

Thank you.

"""

detect_phishing(sender, subject, body)
```

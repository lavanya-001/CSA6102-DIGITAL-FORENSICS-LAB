```python
# DNS Lookup Program using socket module

import socket

# Function to perform DNS lookup
def dns_lookup(domain):
    try:
        # Get IP address of the domain
        ip = socket.gethostbyname(domain)

        print("===== DNS Lookup Result =====")
        print("Domain Name :", domain)
        print("IP Address  :", ip)

    except socket.gaierror:
        print("DNS lookup failed! Invalid domain name.")

# Take domain name from the user
domain = input("Enter the domain name (e.g., google.com): ")

# Call the function
dns_lookup(domain)
```

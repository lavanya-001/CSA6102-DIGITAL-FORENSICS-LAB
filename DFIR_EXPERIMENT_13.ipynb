```python
# Digital Forensic Report Generator

case_id = input("Enter Case ID: ")
investigator = input("Enter Investigator Name: ")
date = input("Enter Date: ")
evidence = input("Enter Evidence Collected: ")
finding = input("Enter Investigation Findings: ")
conclusion = input("Enter Conclusion: ")

# Create report
report = f"""
DIGITAL FORENSIC REPORT
============================

Case ID           : {case_id}
Investigator      : {investigator}
Date              : {date}

Evidence Collected:
-------------------
{evidence}

Investigation Findings:
-----------------------
{finding}

Conclusion:
-----------
{conclusion}

============================
End of Report
"""

# Display report
print(report)

# Save report to a file
with open("Digital_Forensic_Report.txt", "w") as f:
    f.write(report)

print("Report saved as Digital_Forensic_Report.txt")
```

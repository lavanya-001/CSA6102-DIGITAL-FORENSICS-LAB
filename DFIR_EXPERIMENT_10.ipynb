```python
from google.colab import files
import re

# Upload email header file
uploaded = files.upload()

filename = list(uploaded.keys())[0]

# Read the email header
with open(filename, "r") as f:
    header = f.read()

# Extract fields
from_match = re.search(r"From:\s.*<(.+?)>", header)
return_match = re.search(r"Return-Path:\s*<(.+?)>", header)
received = re.findall(r"Received:.*", header)

print("\nEmail Header Analysis")
print("-" * 30)

if from_match:
    from_email = from_match.group(1)
    print("From:", from_email)
else:
    from_email = None
    print("From: Not Found")

if return_match:
    return_email = return_match.group(1)
    print("Return-Path:", return_email)
else:
    return_email = None
    print("Return-Path: Not Found")

print("\nReceived Headers:")
for r in received:
    print(r)

# Basic spoof detection
if from_email and return_email:
    from_domain = from_email.split("@")[-1]
    return_domain = return_email.split("@")[-1]

    print("\nSpoofing Check")
    print("-" * 20)

    if from_domain != return_domain:
        print("⚠ Possible Spoofed Sender Detected!")
        print("From Domain      :", from_domain)
        print("Return-Path Domain:", return_domain)
    else:
        print("No Spoofing Indicators Found.")
```
